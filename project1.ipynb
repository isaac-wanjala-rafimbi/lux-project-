# Statistical Summary
# Summary of Statistical Data:
The dataset comprises six main numerical features: Temperature (Temp_C), Dew Point Temperature (Dew Point Temp_C), Relative Humidity (Rel Hum_%), Wind Speed (Wind Speed_km/h), Visibility (Visibility_km), and Pressure (Press_kPa). The average Temperature is 8.80°C with a standard deviation of 11.69°C, indicating considerable fluctuation. Dew Point Temperature has a lower average of 2.56°C but exhibits similar variability. Relative Humidity averages 67.43%, Wind Speed averages 14.95 km/h, Visibility is generally high with an average of 27.66 km, and Pressure remains steady with an average of 101.05 kPa, showing minimal fluctuation.

# Outliers: Identification and Analysis:
The dataset highlights several potential outliers, particularly in Temperature and Dew Point Temperature. Notably, the minimum Temperature of -23.3°C and Dew Point Temperature of -28.5°C are much lower than the average, possibly reflecting abnormal weather conditions. Additionally, a Wind Speed of 83 km/h stands out as an outlier compared to the average of 14.95 km/h, suggesting extreme weather events. These outliers may be key in understanding rare weather occurrences or might need further examination to confirm data accuracy.

# correlation matrices and heatmaps to identify relationships between different weather parameters.

```python
# Select only numerical columns
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the Spearman correlation matrix and round to 2 decimal places
correlation_matrix = numerical_df.corr(method='pearson').round(2)

# Style the correlation matrix with a uniform green background color
styled_matrix = correlation_matrix.style.background_gradient(axis=None, cmap='Blues')

# Display the styled correlation matrix
styled_matrix
```

```python
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr().round(2)

# Create a figure with a size of 8x8
plt.figure(figsize=(8, 8))

# Create a heatmap
sns.heatmap(correlation_matrix, 
            annot=True,              # Annotate with correlation values
            cmap='Blues',           # Use green colormap
            center=0,                # Center the colormap at 0
            vmin=-1, vmax=1,         # Set color scale range
            linewidths=0.5,         # Line width between cells
            linecolor='black')      # Line color between cells

# Add labels and title
plt.title('Heatmap of Correlation Matrix')  # Plot title

# Display the plot
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

Patterns and Trends in the Data:

Temperature Variations:

Monthly Patterns: January and February are the coldest months, with average temperatures around -7.37°C and -4.23°C, respectively. July is the warmest month, with an average temperature of about 22.79°C. Temperatures begin to rise in March, peak in July, and start to decline in September.

Seasonal Patterns:

Winter (December to February): The average temperature is around -4.98°C.
Summer (June to August): The average temperature is 21.75°C.
Spring (March to May) and Autumn (September to November): These seasons have moderate temperatures, ranging from an average of 8.81°C to 9.47°C.
Humidity Variations:

Monthly Patterns: The highest humidity levels are observed in December and October, with average humidity of 81.30% and 75.73%, respectively. April has the lowest humidity, averaging 56.15%.

Seasonal Trends:

Winter: The highest average humidity is observed, at 72.97%.
Spring: The lowest average humidity, at 60.98%.
Summer and Autumn: Humidity levels are moderate, with averages of 63.57% and 72.33%, respectively.
These observations highlight a clear seasonal pattern where temperatures peak in the summer and drop in winter, while humidity levels are generally higher during the colder months.

# 4.2 Investigate any anomalies or unusual patterns in the data. What might be the reasons for these anomalies?

Unusual Patterns and Anomalies:

Temperature Anomalies:

On January 3, 2012, temperatures dropped to an unusually low -17.3°C. Conversely, on August 26, 2012, temperatures soared to a high of 30.4°C.
Humidity Anomalies:

Exceptionally high humidity levels, reaching close to or exactly 100%, were observed on several occasions, particularly on March 14, 2012, and December 2, 2012.
Possible Reasons for These Anomalies:

Temperature Drops in January: The very low temperatures in January may have been caused by a cold spell or arctic front, common in winter, leading to sharp temperature declines.
Temperature Spikes in August: The high temperatures in August could be attributed to a heatwave, which often occurs in late summer, resulting in unusually high temperatures.
High Humidity Levels: The extreme humidity readings might be linked to weather conditions like fog, drizzle, or periods of heavy rainfall, leading to air saturation.
While these anomalies align with typical weather patterns, their extremity may require further investigation into specific weather events or local conditions at those times.

5.2 How can these insights be useful for weather prediction or other practical applications?

Practical Applications of the Insights:

Weather Forecasting:

Seasonal Forecasting: Understanding the predictable seasonal temperature cycles and humidity patterns can enhance the accuracy of seasonal weather forecasts. For example, predicting colder temperatures and higher humidity in winter can aid in preparing for snow or ice events.
Anomaly Detection: By identifying extreme temperature and humidity variations, models can be developed to predict similar anomalies. This can allow for early warnings of heatwaves or cold spells, reducing their impact on infrastructure, agriculture, and public health.
Agriculture:

Crop Planning: Farmers can use insights on temperature and humidity patterns to plan planting and harvesting schedules. For instance, anticipating summer heat can guide the selection of heat-resistant crops, while knowing that spring is typically drier can inform irrigation needs.
Frost Protection: Forecasting cold spells can help farmers take protective measures against frost, such as using frost covers or heaters to safeguard crops.
Public Safety and Infrastructure:

Extreme Weather Preparedness: Authorities can use these insights to prepare for extreme weather events. For example, getting ready for summer heatwaves and ensuring adequate heating resources in winter can mitigate risks for vulnerable populations.
Infrastructure Maintenance: High humidity, particularly in autumn and winter, can accelerate the wear and tear of buildings and roads. Maintenance schedules can be adjusted based on these insights to prolong the lifespan of infrastructure.
Logistics and Transportation:

Visibility and Safety: High humidity levels in winter and autumn, associated with foggy conditions, can impact visibility on roads and at airports. By anticipating when these conditions are likely to occur, safety can be improved by increasing weather-related alerts or adjusting speed limits.
Supply Chain Management: Anticipating extreme weather events can assist in planning logistics and supply chains, ensuring deliveries are made before conditions deteriorate or adjusting routes to avoid affected areas.
